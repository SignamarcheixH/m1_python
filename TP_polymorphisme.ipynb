````python
class CompteBancaire :
    def __init__(self, nom="Dupont", solde=1000) :
        self._nom=nom
        self._solde=solde
    
    def _get_solde(self):
        return self._solde
    
    def _set_solde(self, newSolde):
        self._solde = newSolde
    
    def depot(self, somme):
        self._solde+=somme
        
    def retrait(self, somme):
        if(self._solde >= somme):
            self._solde-=somme
        else:
            print("Retrait de {} interdit : Fonds insuffisant".format(somme))
            
    def affiche(self):
        print("Le solde du compte bancaire de {} est de {}".format(self._nom, self._solde))
        
    solde=property(_get_solde, _set_solde)

compte1=CompteBancaire('Hugues', 2000)
compte1.depot(500)
compte1.retrait(200)
compte1.affiche()
compte2=CompteBancaire('Denis', 400)
compte2.retrait(500)
compte2.affiche()
````

````python
class Point:
    def __init__(self, coords={}):
        self._coords={}
        for coord in coords.items():
            self._coords[coord[0]] = coord[1]
            
    def toString(self):
        values = ''
        for value in self._coords.values():
            values=values+str(value)+', '
        values=values[0:-2]
        print("P({})".format(values))
        
p1=Point({'x':1, 'y':2})
p1.toString()

p2=Point({'x':2, 'y':3, 'z':56, 'tr':23})
p2.toString()

        
````

````python

class DateNaissance:
    def __init__(self, jour, mois, annee):
        self._jour=jour 
        self._mois=mois
        self._annee=annee
    
    def _get_jour(self):
        return self._jour
    
    def _get_mois(self):
        return self._mois
    
    def _get_annee(self):
        return self._annee
    
    def _set_jour(self, newJour):
        self._jour=newJour
        
    def _set_mois(self, newMois):
        self._mois=newMois
        
    def _set_annee(self, newAnnee):
        self._annee=newAnnee
        
    def toString(self):
        return("{} / {} / {}".format(self._jour, self._mois, self._annee))
    
    jour=property(_get_jour, _set_jour)
    mois=property(_get_mois, _set_mois)
    annee=property(_get_annee, _set_annee)
    

class Personne:
    def __init__(self, nom, prenom, datenaissance):
        self._nom = nom
        self._prenom = nom
        self._datenaissance = datenaissance
        
    def afficher(self):
        print("Nom : {}".format(self._nom))
        print("Prénom : {}".format(self._nom))
        print("Date de Naissance : {}".format(self._datenaissance.toString()))
        
class Employe(Personne):
    def __init__(self, nom, prenom, datenaissance, salaire):
        Personne.__init__(self, nom, prenom, datenaissance)
        self._salaire = salaire
        
    def afficher(self):
        Personne.afficher(self)
        print("Salaire : {}".format(self._salaire))
        

class Chef(Employe):
    def __init__(self, nom, prenom, datenaissance, salaire, service):
        Employe.__init__(self, nom, prenom, datenaissance, salaire)
        self._service = service
        
    def afficher(self):
        Employe.afficher(self)
        print("Service : {}".format(self._service))
    
naissanceHugues = DateNaissance(1,4,1997)
naissanceHugues.toString()
hugues=Personne('Signamarcheix', 'Hugues', naissanceHugues)
hugues.afficher()

print("")

albert=Employe('Camus', 'Albert', DateNaissance(8,2,1900), 1500)
albert.afficher()

print("")

sylvie=Chef("Sylvie","Tellier", DateNaissance(1,7,1988),7865.548,"Ressource humaine")
sylvie.afficher()
````

## TP :  Le polymorphisme

### Consignes 

Une boîtes aux lettres recueille des courrier qui peuvent être des lettres ou des colis.

Une lettre est caractérisée par :

son poids (en grammes)
le mode d’expédition (express ou normal)
son adresse de destination
son adresse d’expédition
son format (“A3” ou “A4”)
Un colis est caractérisé par :

son poids (en grammes)
le mode d’expédition (express ou normal)
son adresse de destination
son adresse d’expédition
son volume (en litres )
Il faut donc définir aussi les deux méthodes:

calculTimbre: qui calcule le prix du Timbre
ToString qui affiche un courrier
Règles pour calculer le prix du Timbre :

en mode d’expédition normal :

Le montant nécessaire pour affranchir une lettre dépend de son format et de son poids : Formule : montant = tarif de base + 1.0 * poids (kilos), où le tarif de base pour une lettre “A4” est de 2.50, et 3.50 pour une lettre “A3”

Le montant nécessaire pour affranchir un colis dépend de son poids et de son volume : Formule : montant = 0.25 volume (litres) + poids (kilos) 1.0

en mode d’expédition express : les montants précédents sont doublés, quelque soit le type de courrier

### Exemple d’exécution :
```
>>>L1=Lettre(“Lille”,”Paris”,80,”normal”,”A4″)
>>>L1.ToString()
Lettre:
Adresse destination: Lille 
Adress expedition: Paris
Poids: 80.00 grammes
Mode: normal 
Format:A4
Prix du timbre:0.20
>>>C1=Colis(“Marrakeche”,”Barcelone “,3500,”expresse”,2.25)
>>>C1.ToString()
Collis:
Adresse destination: Marrakeche 
Adress expedition: Barcelone 
Poids: 3500.00 grammes
Mode: expresse 
Volume: 2.25 litres
Prix du timbre:3937.50
```

````python
class Courrier:
    EXPEDITION_MODES=["EXPRESS", "NORMAL"]
    def __init__(self, poids, mode_expedition, adresse_destination, adresse_expedition):
        assert mode_expedition in self.EXPEDITION_MODES, 'mode_expedition doit être EXPRESS ou NORMAL'
        assert poids >= 1,'le poids doit être supérieur ou égal à 1'
        self._poids = poids
        self._mode_expedition = mode_expedition
        self._adresse_destination = adresse_destination
        self._adresse_expedition = adresse_expedition
        
    def toString(self):
        print("Adresse destination : {}".format(self._adresse_destination))
        print("Adresse expedition : {}".format(self._adresse_expedition))
        print("Poids : {}".format(self._poids))
        print("Mode : {}".format(self._mode_expedition))
        

class Lettre(Courrier):
    LETTER_FORMATS=["A3", "A4"]
    def __init__(self, poids, mode_expedition, adresse_destination, adresse_expedition, lettre_format):
        assert lettre_format in self.LETTER_FORMATS, 'lettre_format doit être A3 ou A4'
        Courrier.__init__(self, poids, mode_expedition, adresse_destination, adresse_expedition)
        self._lettre_format = lettre_format
        
    def calculTimbre(self):
        tarif_base = 3.50 if self._lettre_format=="A3" else 2.50
        tarif = tarif_base * (self._poids/1000)
        return tarif if self._mode_expedition=="NORMAL" else tarif*2
        
    def toString(self):
        print("Lettre :")
        Courrier.toString(self)
        print("Format : {}".format(self._lettre_format))
        print("Prix du timbre : {}".format(self.calculTimbre()))
    
    
class Colis(Courrier):
    def __init__(self, poids, mode_expedition, adresse_destination, adresse_expedition, volume):
        assert volume >= 1, 'le volume doit être supérieur ou égal à 1' 
        Courrier.__init__(self, poids, mode_expedition, adresse_destination, adresse_expedition)
        self._volume = volume
    
    def toString(self):
        print("Colis :")
        Courrier.toString(self)
        print("Volume : {}".format(self._volume))
        print("Prix du timbre : {}".format(self.calculTimbre()))
    
    def calculTimbre(self):
        tarif = round((0.25 * self._volume) * (self._poids/1000),2)
        return tarif if self._mode_expedition == "NORMAL" else tarif*2

    
    
lettre=Lettre(80, "NORMAL", "Paris", "Lyon", "A4")
colis=Colis(3500, "EXPRESS", "Marseille", "Pau", 2.25)
lettre.toString()
print("")
colis.toString()




````
